# Logisitc Regression Model

## Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Load the preprocessed data

```python
df_preprocessed = pd.read_csv('data/processed/Laliga_fixtures_preprocessed.csv')
```

```python
df_preprocessed.head()
```

```python
df_preprocessed.describe()
```

## Exploring the targets

```python
df_preprocessed['Target'].value_counts()
```

```python
df_preprocessed['Target'].sum() / len(df_preprocessed['Target'])
```

## Selecting Inputs

```python
targets = df_preprocessed.iloc[:, -1]
```

```python
unscaled_inputs = df_preprocessed.iloc[:, :-1]
unscaled_inputs = unscaled_inputs.drop(columns = ['G_Home', 'G_Away'])
unscaled_inputs.head(5)
```

```python
unscaled_inputs.columns.values
```

```python
columns_as_inputs = ['Time_Sin', 'Time_Cos', 'Wk', 'Home Position',
       'Home_code_0', 'Home_code_1', 'Home_code_2', 'Home_code_3',
       'Home_code_4', 'xG_Home', 'xG_Away', 'Away_code_0', 'Away_code_1',
       'Away_code_2', 'Away_code_3', 'Away_code_4', 'Away Position']
```

```python
unscaled_inputs = unscaled_inputs[columns_as_inputs]
```

## Standardization

```python
columns_to_standardize = ['xG_Home', 'xG_Away']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
unscaled_inputs[columns_to_standardize] = scaler.fit_transform(unscaled_inputs[columns_to_standardize])
```

```python
scaled_inputs = unscaled_inputs.copy()
```

```python
scaled_inputs
```

```python
scaled_inputs.shape
```

```python
scaled_inputs.columns.values
```

## Splitting the data into training and testing, and shuffling

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, targets, train_size=0.8, random_state=20)
```

```python
print(x_train.shape, y_train.shape)
```

```python
print(x_test.shape, y_test.shape)
```

## Logistic Regression with SK-Learn

### Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg = LogisticRegression(max_iter=1000)
```

```python
reg.fit(x_train, y_train)
```

```python
reg.score(x_train, y_train)
```

### Intercepts and Coefficients

```python
reg.intercept_
```

```python
reg.coef_
```

```python
feature_name = unscaled_inputs.columns.values
```

```python
summary_table = pd.DataFrame(columns=['Feature name'], data = feature_name)

summary_table['Coefficient'] = np.transpose(reg.coef_)

summary_table
```

```python
summary_table.index = summary_table.index + 1
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table
```

```python
summary_table['Odds_ratio'] = np.exp(summary_table.Coefficient)
```

```python
summary_table
```

```python
summary_table.sort_values('Odds_ratio', ascending=False)
```

### Testing

```python
reg.score(x_test, y_test)
```

```python
predict_proba = reg.predict_proba(x_test)
predict_proba
```

```python
predict_proba.shape
```

```python
predict_proba[:, 1]
```
