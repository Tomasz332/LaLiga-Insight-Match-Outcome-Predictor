```python
import pandas as pd
import numpy as np
```

```python
year = 2025
```

```python
URL = f'https://fbref.com/en/comps/12/{year}-{year+1}/schedule/{year}-{year+1}-La-Liga-Scores-and-Fixtures'
TABLE_ID = f'sched_{year}-{year+1}_12_1'
```

```python
df = pd.read_html(URL
                       , attrs={"id": TABLE_ID})[0]
```

```python
df
```

```python
df = df.dropna(subset=['Wk'])
```

```python
df = df.dropna(subset=['Score'])
```

```python
df
```

```python
df.columns.values
```

```python
final_df = df[['Wk', 'Time', 'Home', 'xG', 'Score', 'xG.1', 'Away']]
```

```python
final_df.head()
```

```python
fix_2025_df = pd.DataFrame()
```

```python
fix_2025_df['Matchweek'] = final_df['Wk']
fix_2025_df['Time'] = final_df['Time'] 
fix_2025_df['Home'] = final_df['Home'] 
fix_2025_df['xG_Home'] = final_df['xG']
fix_2025_df['xG_Away'] = final_df['xG.1']
fix_2025_df['Away'] = final_df['Away']
```

```python
fix_2025_df.head()
```

```python
actual_scores = final_df['Score']
```

```python
fix_2025_df.to_csv('2025-season-fix.csv')
actual_scores.to_csv('2025-season-fix-actual-score.csv')
```
