```python
import pandas as pd
import numpy as np
import os 
```

```python
year = 2017
```

```python
URL = f'https://fbref.com/en/comps/12/{year}-{year+1}/schedule/{year}-{year+1}-La-Liga-Scores-and-Fixtures'
TABLE_ID = f'sched_{year}-{year+1}_12_1'
FILENAME = 'LaLigaFixturesFBREF.csv'
```

```python
df = pd.read_html(URL
                       , attrs={"id": TABLE_ID})[0]
```

```python
df.head(5)
```

```python
df['Wk'].isnull().sum()
```

```python
df = df.dropna(subset=['Wk'])
```

```python
final_df = df.copy()
```

```python
final_df['Season'] = f"{year}-{year + 1}"
```

```python
final_df
```

```python
final_df.columns
```

```python
final_df = final_df[['Season','Wk', 'Day', 'Date', 'Time', 'Home', 'xG', 'Score', 'xG.1', 'Away',
       'Attendance', 'Venue', 'Referee']]
```

```python
final_df
```

```python
if not os.path.exists(FILENAME):
    final_df.to_csv(FILENAME, index=False)
else:
    final_df.to_csv(FILENAME, mode='a', header=False, index=False)
```
